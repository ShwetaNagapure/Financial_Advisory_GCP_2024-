```python
import base64
import vertexai
from vertexai.generative_models import GenerativeModel, Part, FinishReason
import vertexai.preview.generative_models as generative_models

def generate():
  vertexai.init(project="bob-rg7-gcp", location="us-central1")
  model = GenerativeModel(
    "gemini-1.5-pro-001",
  )
  responses = model.generate_content(
      [prompt_template, document1, """User-provided question: {data_dict[\'question\']}
Text and/or tables: {formatted_texts}"""],
      generation_config=generation_config,
      safety_settings=safety_settings,
      stream=True,
  )

  for response in responses:
    print(response.text, end="")


document1 = Part.from_data(
    mime_type="application/pdf",
    data=base64.b64decode(""""""))
prompt_template = """You are an AI assistant that helps people with financial advice. The answer should be descriptive.
I am of age: {age}
occupation: {occupation}
monthly income: {monthly_income}
monthly expenses: {monthly_expenses}
current saving: {current_saving}
investment goals: {investment_goals}
risk tolerance: {risk_tolerance}
car price is {car_price}
loan term: {loan_term}
on floating interest rate

What is the interest rate being offered on the loan? {interest_rate}
Do you have any other outstanding loans? {outstanding_loans}
What is your credit score? {credit_score}

Please provide me with car loan advice."""

# Function to get user input
def get_user_data():
    user_data = {
        "age": input("Enter your age: "),
        "occupation": input("Enter your occupation: "),
        "monthly_income": input("Enter your monthly income: "),
        "monthly_expenses": input("Enter your monthly expenses: "),
        "current_saving": input("Enter your current saving: "),
        "investment_goals": input("Enter your investment goals: "),
        "risk_tolerance": input("Enter your risk tolerance: "),
        "car_price": input("Enter the car price: "),
        "loan_term": input("Enter the loan term: "),
        "interest_rate": input("Enter the interest rate being offered on the loan: "),
        "outstanding_loans": input("Do you have any other outstanding loans? "),
        "credit_score": input("Enter your credit score: ")
    }
    return user_data

# Function to generate the prompt based on user data
def create_prompt(user_data):
    return prompt_template.format(
        age=user_data["age"],
        occupation=user_data["occupation"],
        monthly_income=user_data["monthly_income"],
        monthly_expenses=user_data["monthly_expenses"],
        current_saving=user_data["current_saving"],
        investment_goals=user_data["investment_goals"],
        risk_tolerance=user_data["risk_tolerance"],
        car_price=user_data["car_price"],
        loan_term=user_data["loan_term"],
        interest_rate=user_data["interest_rate"],
        outstanding_loans=user_data["outstanding_loans"],
        credit_score=user_data["credit_score"]
    )

# Generation configuration
generation_config = {
    "max_output_tokens": 2192,
    "temperature": 0.95,
    "top_p": 0.75,
}

# Safety settings
safety_settings = {
    generative_models.HarmCategory.HARM_CATEGORY_HATE_SPEECH: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_HARASSMENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
}

# Function to generate response using Vertex AI
def generate(prompt, generation_config, safety_settings):
    vertexai.init(project="bob-rg7-gcp", location="us-central1")
    model = generative_models.GenerativeModel("gemini-1.5-pro-001")

    responses = model.generate_content(
        [prompt],
        generation_config=generation_config,
        safety_settings=safety_settings,
        stream=True,
    )

    for response in responses:
        print(response.text, end="")

# Main code execution
user_data = get_user_data()
prompt = create_prompt(user_data)
generate(prompt, generation_config, safety_settings)
```

```python

import ipywidgets as widgets
from IPython.display import display, HTML
import base64
import vertexai
from vertexai.generative_models import GenerativeModel, Part
import vertexai.preview.generative_models as generative_models
from google.colab import auth, files

# Authenticate user
auth.authenticate_user()

pdf_filename ="/content/carloan.pdf"

with open(pdf_filename, 'rb') as file:
    pdf_content = file.read()

document1 = Part.from_data(
    mime_type="application/pdf",
    data=base64.b64encode(pdf_content).decode()
)

prompt_template = """You are an AI assistant that helps people with financial advice. The answer should be descriptive.
I am of age: {age}
occupation: {occupation}
monthly income: {monthly_income}
monthly expenses: {monthly_expenses}
current saving: {current_saving}
investment goals: {investment_goals}
risk tolerance: {risk_tolerance}
car price is {car_price}
loan term: {loan_term}
on floating interest rate

What is the interest rate being offered on the loan? {interest_rate}
Do you have any other outstanding loans? {outstanding_loans}
What is your credit score? {credit_score}

Please provide me with {query} advice."""

generation_config = {
    "max_output_tokens": 1024,
    "temperature": 0.95,
    "top_p": 0.75,
}

safety_settings = {
    generative_models.HarmCategory.HARM_CATEGORY_HATE_SPEECH: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_DANGEROUS_CONTENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_SEXUALLY_EXPLICIT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
    generative_models.HarmCategory.HARM_CATEGORY_HARASSMENT: generative_models.HarmBlockThreshold.BLOCK_MEDIUM_AND_ABOVE,
}

# Create input widgets
age = widgets.IntText(description='Age:')
occupation = widgets.Text(description='Occupation:')
monthly_income = widgets.FloatText(description='Monthly Income:')
monthly_expenses = widgets.FloatText(description='Monthly Expenses:')
current_saving = widgets.FloatText(description='Current Savings:')
investment_goals = widgets.Text(description='Investment Goals:')
risk_tolerance = widgets.Text(description='Risk Tolerance:')
car_price = widgets.FloatText(description='Car Price:')
loan_term = widgets.Text(description='Loan Term:')
interest_rate = widgets.FloatText(description='Interest Rate:')
outstanding_loans = widgets.Text(description='Outstanding Loans:')
credit_score = widgets.IntText(description='Credit Score:')
query = widgets.Text(description='Query:')

output = widgets.Output()

def get_advice(b):
    with output:
        output.clear_output()
        print("Getting advice...")

        # Prepare the data
        data = {
            'age': age.value,
            'occupation': occupation.value,
            'monthly_income': monthly_income.value,
            'monthly_expenses': monthly_expenses.value,
            'current_saving': current_saving.value,
            'investment_goals': investment_goals.value,
            'risk_tolerance': risk_tolerance.value,
            'car_price': car_price.value,
            'loan_term': loan_term.value,
            'interest_rate': interest_rate.value,
            'outstanding_loans': outstanding_loans.value,
            'credit_score': credit_score.value,
            'query': query.value
        }

        # Generate the prompt
        prompt = prompt_template.format(**data)

        # Initialize Vertex AI
        vertexai.init(project="bob-rg7-gcp", location="us-central1")
        model = GenerativeModel("gemini-1.5-pro-001")

        # Generate the response
        response = model.generate_content(
            [prompt, document1],
            generation_config=generation_config,
            safety_settings=safety_settings,
        )

        print("Advice:")
        print(response.text)

# Create a button
button = widgets.Button(description="Get Advice")
button.on_click(get_advice)

# Display the form
form = widgets.VBox([age, occupation, monthly_income, monthly_expenses, current_saving,
                     investment_goals, risk_tolerance, car_price, loan_term,
                     interest_rate, outstanding_loans, credit_score, query, button, output])

display(HTML("<h1>Financial Advice Chatbot</h1>"))
display(form)
```
